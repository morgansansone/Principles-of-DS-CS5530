### Import Libraries & Add Data to DF

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/morgansansone/Principles-of-DS-CS5530/refs/heads/main/Assignment2_3/Question1/raw_data/train.csv')
print(df.shape)
```

###1) Find missing values in all columns and replace or drop them

```python
# missing counts & percentages
#df.isna().sum() # counts
(df.isna().mean() * 100).round(2) # percentages
```

```python
# drop New_Price since it is missing 86% of its values
df_cars = df.copy()
df_cars.drop(columns=['New_Price'], inplace=True)
```

I chose to drop New_Price because it was missing 86.06% of its values which is more that 75%. That shows that theres too little of values to be able to fill all the NaN values with a median or mean so I chose to drop the overall column. I know the next question says to strip it but its mostly empty volumn so it wouldnt mean much to keep using it.

```python
df_cars.isna().sum()
```

###2) Remove units from attributes and keep numerical values


```python
# remove units from Mileage (kmpl), Engine (CC), Power (bhp)
df_cars['Mileage'] = df_cars['Mileage'].astype(str).str.extract(r'([-+]?\d*\.?\d+)')[0].astype(float)
df_cars['Engine'] = df_cars['Engine'].astype(str).str.extract(r'([-+]?\d*\.?\d+)')[0].astype(float)
df_cars['Power'] = df_cars['Power'].astype(str).str.extract(r'([-+]?\d*\.?\d+)')[0].astype(float)
```

```python
df_cars.fillna(df_cars.median(numeric_only=True), inplace=True) # filling na values (now that they are numeric)
df_cars.isna().sum() # checking again
```

I used these values medians because of the spectrum of years and power for the cars. The categorical ones would have had a mode if they were missing any, but thankfully it was only numerical variables.

###3) Change categorical variables into numerical one hot encoded values

```python
df_cars = pd.get_dummies(df_cars, columns=['Fuel_Type', 'Transmission'], dtype=int)
```

```python
df_cars.head(5) #checking
```

###4) Creation of new feature with the added column (mutate)

```python
current_year = datetime.now().year # get current year
df_cars['Car_Age'] = current_year - df_cars['Year'] # new column
df_cars.to_csv('cars_cleaned.csv', index=False) # save cleaned data
```

```python
df_cars.head(5) #checking
```

###5) Select, filter, rename, mutate, arrange, and summarize with group by operations

```python
# select columns
selected = df_cars[['Name', 'Location', 'Car_Age', 'Kilometers_Driven', 'Price']]
selected.head(5)
```

```python
# filter Car_Age > 10
filtered = df_cars[df_cars['Car_Age'] > 10]
filtered.head(5) #checking
df_cars.to_csv('cars_cleaned.csv', index=False) #save to csv cleaned data
```

```python
# rename engine, price, and power
renamed = df_cars.rename(columns={'Engine': 'Engine_CC', 'Price': 'Price_Lakh', 'Power': 'Power_bhp'})
renamed.columns.tolist()
```

```python
# mutate price per km
df_cars['Price_per_KM'] = df_cars['Price'] / df_cars['Kilometers_Driven']
df_cars[['Name', 'Price', 'Kilometers_Driven', 'Price_per_KM']].head()
```

```python
# arrange price descending
arranged = df_cars.sort_values('Price', ascending=False)
arranged[['Name', 'Price', 'Car_Age']].head()
```

```python
# summarize avg price and mileage/location
summary = (df_cars.groupby('Location')
                  .agg(avg_price=('Price', 'mean'),
                       avg_mileage=('Mileage', 'mean'),
                       n_cars=('Name', 'count'))
                  .reset_index())
summary
```
